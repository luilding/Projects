```python
import requests
import folium
from geopy.geocoders import Nominatim
from IPython.display import display
import openrouteservice

def geocode_address(address):
    geolocator = Nominatim(user_agent="your_app_name")
    location = geolocator.geocode(address)
    if location:
        return location.latitude, location.longitude
    else:
        print("Address not found")
        return None

def visualize_location(lat, lng):
    #Create map centered at the location
    m = folium.Map(location=[lat, lng], zoom_start=15)
    #Add a marker at the location
    folium.Marker([lat, lng], popup=f"Latitude: {lat}, Longitude: {lng}").add_to(m)
    #Display
    display(m)

def get_route(start_lat, start_lng, end_lat, end_lng, api_key):
    client = openrouteservice.Client(key=api_key)
    coordinates = [[start_lng, start_lat], [end_lng, end_lat]]
    route = client.directions(coordinates=coordinates, profile='driving-car', format='geojson')
    return route

# Function to visualize the route on a map
def visualize_route(route):
    m = folium.Map(location=[route['features'][0]['geometry']['coordinates'][0][1], 
                             route['features'][0]['geometry']['coordinates'][0][0]], zoom_start=13)
    
    folium.GeoJson(route, name="route").add_to(m)
    folium.LayerControl().add_to(m)
    display(m)

user_input = input("Please enter your current location: ")
location = geocode_address(user_input)
if location:
    print(f"Geocoded location: Latitude = {location[0]}, Longitude = {location[1]}")
    visualize_location(location[0], location[1])

    destination_input = input("Please enter your destination: ")
    destination = geocode_address(destination_input)
    if destination:
        print(f"Geocoded destination: Latitude = {destination[0]}, Longitude = {destination[1]}")
        
        # Get the route
        api_key = '5b3ce3597851110001cf6248a6b7c97bb850491794bb504b30e2f2f7'
        route = get_route(location[0], location[1], destination[0], destination[1], api_key)
        
        # Visualize the route
        visualize_route(route)
    else:
        print("Destination location could not be determined.")
else:
    print("Current location could not be determined.")
```
